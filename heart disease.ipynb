```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams
from matplotlib.cm import rainbow
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('heart.csv')
```

```python
df.describe()
```

```python
#check correlations of each features in dataset
corrmat = df.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(20,20))
g=sns.heatmap(df[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

```python
#get dummy variables
dataset = pd.get_dummies(df, columns = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal'])
```

```python
standardScaler = StandardScaler()
columns_to_scale = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
dataset[columns_to_scale] = standardScaler.fit_transform(dataset[columns_to_scale])
```

```python
dataset.head(5)
```

```python
y = dataset['target']
x = dataset.drop(['target'], axis = 1)
```

```python
#classification using KNN model
knn_classifier = KNeighborsClassifier(n_neighbors = 10)
score1=cross_val_score(knn_classifier,x,y,cv=10)
```

```python
score1.mean()
```

```python
#classification using random forest model
randomforest_classifier= RandomForestClassifier(n_estimators=10)
score2=cross_val_score(randomforest_classifier,x,y,cv=10)
```

```python
score2.mean()
```

```python
#classification using decision tree model
decisiontree=DecisionTreeClassifier(random_state=0)
score3=cross_val_score(decisiontree,x,y,cv=10)
```

```python
score3.mean()
```
